```python
import matplotlib.pyplot as  plt
import numpy as np
x = np.linspace(0,10,num=1000)
x
```

```python
y = np.sin(x)
plt.plot(x, y, label="y=sinx",c='red', linewidth=4,marker='v',markersize=4,markeredgecolor='black',markerfacecolor='blue')
plt.xlabel('Times')
plt.ylabel("vol")
plt.title('this is line')
plt.legend(loc='right')
plt.show()
```

```python
x1 = np.linspace(0,10,100)
```

```python
y1 = np.sin(x1)
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.plot(x1,y1,ls='-',lw=2,label='x和y的关系')
plt.legend(loc='upper center')
```

```python
x2 = np.arange(0,1.1,0.01)
y2 = x2 **2
```

```python
plt.figure(figsize=(6.4,4.84),dpi=200,facecolor='blue')
plt.title('这是个标题')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xticks([0,0.2,0.4,0.6,0.8,1])
plt.yticks([0,0.2,0.4,0.6,0.8,1])
plt.plot(x2,y2,label="y=x^2")
plt.legend(loc='best')

plt.xlabel('x轴')
plt.ylabel('y轴')
plt.savefig('可视化.png')
plt.show()
```
